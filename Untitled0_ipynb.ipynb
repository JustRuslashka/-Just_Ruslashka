```python
# 1.загрузим данные
import pandas as pd
df=pd.read_csv("iris.csv")
```

```python
# 2.выводим первые 5 строчек файла
df.head(5)
```

```python
# 3.1.воспользуемся командами groupBy, max, 
#   чтобы вывести максимальное значние каждого из аттрбутов цветка.
df[['variety','sepal.length','sepal.width','petal.length','petal.width']].groupby(['variety']).max()
```

```python
# 3.2.воспользуемся командами groupBy, min, 
#   чтобы вывести минимальное значние каждого из аттрбутов цветка.
df[['variety','sepal.length','sepal.width','petal.length','petal.width']].groupby(['variety']).min()
```

```python
# 3.3.воспользуемся командами groupBy, mean, 
#   чтобы вывести среднее значние каждого из аттрбутов цветка.
df[['variety','sepal.length','sepal.width','petal.length','petal.width']].groupby(['variety']).mean()
```

```python
# 4.Постройте точечную диаграмму для petal_length и petal_width,  
#   раскрашенный по типам цветков
import matplotlib.pyplot as plt
import seaborn as sns
sns.lmplot( x="petal.length", y="petal.width", data=df, fit_reg=False, hue='variety', legend=False)
plt.legend(loc='lower right')
```
